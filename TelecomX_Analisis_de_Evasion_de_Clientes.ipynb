```python
import pandas as pd
import requests
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
response = requests.get(url)
datos = response.json()
```

```python
df = pd.DataFrame(datos)
df
```

```python
import pandas as pd
#normalizacion de datos
df_plano = pd.json_normalize(datos)
df_plano
```

```python
df_plano.info()
```

```python
#valores unicos.
for col in df_plano.columns:
  print(f'Valor unico en la columna: "{col}" : {df_plano[col].nunique()}')
  if df_plano[col].nunique() < 30: #imprimir los valores menores de 30
    print(df_plano[col].unique())
    print('-' * 60) #linea divisoria para mejorar la vista
```

```python
#datos duplicados
df_plano.duplicated().sum()
print('Numero de duplicados', df_plano.duplicated().sum())
```

```python
#datos nulos
df_plano.isnull().sum()
print('Numero de nulos', df_plano.isnull().sum())
```

```python
#datos vacios o en blanco
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#cambiamos la columna  account.Charges.Total a tipo de dato float
df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
print(df_plano['account.Charges.Total'].dtype)
```
